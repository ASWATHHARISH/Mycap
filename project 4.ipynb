```python
def make_dict(x):
    dic = {}
    for letter in x:
        dic[letter] = 1 + dic.get(letter, 0)
    return dic
def most_frequent(txt):
    letters = [letter.lower() for letter in txt if letter.isalpha()]
    dic = make_dict(letters)
    result = []
    for key in dic:
        result.append((dic[key], key))
    result.sort(reverse=True)
    for count, letter in result:
        print (letter, count)

t = input("")
most_frequent(t)
```
